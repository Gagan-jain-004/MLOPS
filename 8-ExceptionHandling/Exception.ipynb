``` Exceptions are 
events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:
    \-ZeroDivisionError: Dividing by zero
    \-FileNotFoundError: File not found
    \-ValueError: Invalid value
    \-TypeError: Invalid type 
    ```

````python
## Exception   try,except block      like try catch in cpp

try:
    a=b
except:
    print("The varaible b is not assigned")  # so instead of program crashing it will print this message
````

````python
try:
    result=10/0
except Exception as ex:
    print(ex)  # will print division by zero error message
    print("Please enter a denominator other than zero")
````

````python
## try except else , finally  block   , else will run only if try block is successful without any error
## finally will run always no matter what

try:
    num=int(input("Enter a number: "))
    result=10/num
except ValueError:
    print("Invalid input! Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero! Please enter a non-zero integer.")
except Exception as ex:
    print(ex)    
else:                     # will run if no exception occurs
    print(f"Result is {result}")
finally:                 # will run always
    print("Execution completed.")
````
