### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding is a technique used to represent categorical data as numerical data, which is more unique category . For example, if we have a categorical variable "color" iwth three possible values (red,green,blue). we can represnt it one hot encoding as follows.
- 1. Red: [1,0,0]
- 2. Green: [0,1,0]
- 3. Blue: [0,0,1]

Data encoding is to convert the categorical thing to numerical value as model understands numerical things not human lang 

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a simple data frame 
df= pd.DataFrame({
    'color': ['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## Create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
## perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()    # jo below array hai usko hei sparse matrix bolte hai
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```
