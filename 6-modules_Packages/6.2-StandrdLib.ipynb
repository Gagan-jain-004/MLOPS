```python
## Data Serialization   -  to cnvrt data(in dictionaries etc.) into json,dataframe
import json
data ={'name':'trish','age':41}

# cnvrting json to string
json_str=json.dumps(data)
print(json_str)
print(type(json_str))
#cnvrt string back to actual that it was (dictionry )
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV 

import csv

with open('example.csv',mode="w",newline='')as file:               ## it will create csv file if not exist 
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Trish',78])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
from datetime import datetime,timedelta

now = datetime.now()            # gives current datime
print(now)

yesterday= now-timedelta(days=1)
print(yesterday)
```

```python
## Regular expression  - for matching patterns in strings
import re

pattern=r'\d+'     # pattern to match one or more digits   
text="There are 789 apples and 54 bananas."
match=re.search(pattern,text)
print(match.group())
```
