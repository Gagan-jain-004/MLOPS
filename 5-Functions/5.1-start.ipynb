```python
def evenorodd(num):
    """This function checks if a number is even or odd"""
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
## Call this fxn 
evenorodd(24)
```

```python
# fxn with multiple parameters
def add_numbers(a, b):
    """This function adds two numbers"""
    return a + b

add_numbers(2,5)
```

```python
# Default Parameters
def greet(name="alo"):
    print(f"{name}")
greet()    
```

```python
# Variable length arguments 
## Positional (using * ) ANd keywords arguments (using **)

def printnumbers(*falana):                       # instead of falana standrd is using *args
    for number in falana:
        print(number)

printnumbers(1,2,3,"hell","ken",4,78)       
```

```python
# Positional Argument 
def printnumbers(*args):                    
    for number in args:
        print(number)

printnumbers(1,2,3,"hell","ken",4,78)   
```

```python
# Keywords Arguments (all will be in key value pairs so u can acces like dictionaries)

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="tril",age="78",country="bulgaria")
```

```python
# Combining both positional and keywords arguments 

def printdetails(*args,**kwargs):
    for val in args:
        print(f"Positional: {val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")

printdetails(1,2,3,"SSD",naam="kal",age="78")
```
