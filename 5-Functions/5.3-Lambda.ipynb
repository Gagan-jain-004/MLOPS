LAMDA FUNCTIONS IN PYTHON

Lamda fxns are small anonymous fxns defined using lamda keyword.They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.


 Explaination - anonymous means  without name , one expression means one logical work/task to perform

```python
# Syntax 
lambda arguments: expression
```

```python
# Normally we create fxn as 
def addition(a,b):
    return a+b
addition(2,3)

# convrting to lambda fxn
add = lambda a,b:a+b
type(add)
print(add(2,4))              # passing arguments in variable name
```

```python
# even check
even = lambda x:x%2==0
print(even(20))
```

```python
ad= lambda x,y,z:x+y+z
ad(3,4,5)
```

```python
numbers=[1,2,3]

def square(number):
    return number*number

for num in numbers:
    print(square(num))
    
```

```python
# above is the iterative approach and map is used to eradicate it 
# map syntax        -             map(fxn,iterables)

list(map(lambda x:x**2,numbers))
 
```
