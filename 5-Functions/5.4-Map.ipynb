Map()  fxn 
take input as list or any other iterable and returns a map object (an iterator).

```python
# map syntax        -             map(fxn,iterables)

numbers=[1,2,3]

def square(number):
    return number*number

list(map(square,numbers))
```

```python
#map with lambda

numbers=[1,2,3]
    
list(map(lambda x:x**2,numbers))
```

```python
 ### Map with multiple  iterables 
 numbers1=[1,2,3]
 numbers2=[4,5,6]

 ans=list(map(lambda x,y:x+y,numbers1,numbers2))
 print(ans)
```

```python
## Map to convrt a list of strings to integers
string = ['1','2','3','4']
integer = list(map(int,string))

print(integer)
```

```python
words=['work','like','hell']

uppr = list(map(str.upper,words))
print(uppr)
```

```python
def getname(person):
    return person['name']

people = [
    {'name':"Trish",'age':54},
    {'name':'Jack','age':4}
]

list(map(getname,people))
```
