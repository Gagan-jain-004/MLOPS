```python
# 1 Temperature Conversion

def convrtemp(temp,unit):
    """This Function to convert temp between C and F"""
    if unit=="F":
        temp=(5/9)*(temp - 32)
        return temp
    elif unit=="C":
        temp=((9/5)*temp ) + 32;
        return temp
    else:
        return None

print(convrtemp(5,"C"))        
```

```python
# 2  Password Strength Checker
# - char.isdigit()
# This checks if the current character is a digit (0–9).
# Example: "5".isdigit() → True, "a".isdigit() → False.
# - any(char.isdigit() for char in password)
# The any() function returns True if at least one character in password is a digit.
# If no digits are found, it returns False.

def istrongpass(password):
    """This fxn checks if pass is strong or not """
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return  False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password ):
        return False
    return True
    
#calling 

print(istrongpass("WekPwd"))
print(istrongpass("Str!34skd43sd"))
```

```python
#EXAMPLE 3 : Calculate Total Cost of items in Cart 

## eg cart Data  as List []   of  Dictionaries {}

cart =[
    {'name':'Apple','price':0.5 , 'quantity':4},
    {'name':'banana','price':0.3 , 'quantity':6},
    {'name':'Orange','price':0.7 , 'quantity':3},
]


# fxn to calci total cost 

def calculateTOTcost(cart):
    totalcost=0
    for item in cart:
        totalcost+=item['price']*item['quantity']
    return totalcost

print(calculateTOTcost(cart))
```

```python
# EXAMPLE 4: CHK if string is palindrome or not 

def ispalindrome(s):
    s=s.lower().replace(" ","")    # removing spaces 
    return s==s[::-1]

print(ispalindrome("A man a plan a canal Panama"))
print(ispalindrome("Hell"))    


```

```python
# Factorial using recursion

def fact(n):
    if(n==0):
        return 1
    return n*fact(n-1)

fact(3)
```

```python
# Function To read a FIle and Count the Frequency of Each Word 

def countwordfreq(filepath):
    wordcount={}            # using dictionary ds
    with open(filepath,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('.,!?;:"\'')
                wordcount[word]=wordcount.get(word,0)+1
    return wordcount

filepath ='sample.txt'
countwordfreq(filepath)
```
