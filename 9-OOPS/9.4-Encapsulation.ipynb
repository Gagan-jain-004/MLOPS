## ENCAPSULATION

Concept of bundling data(variables) and methods(functions)  together as a single unit

```python
## Encapsulation with getter and setter methods
## Public , protected, private access modifiers
 

class Person:
    def __init__(self,name,age):
        self.name=name   ## public variables
        self.age=age     

person=Person("Pik",45)
person.name
```

```python
## to create private variables us __ after .

class Person:
    def __init__(self,name,age,gender):
        self.__name=name   ## privat variables
        self.__age=age   
        self.gender=gender  

person=Person("Pik",45,"m")
# person.name            # this wil giv err coz name is private
person.gender
# dir(person)             # to what person has access to 
```

```python
# Protected - cant access outside but can use inside derived class

class Person:
    def __init__(self,name,age,gender):
        self._name=name   ## protected variables
        self._age=age 
        self.gender=gender  

class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)   # calls parent constructor with the inputs

employee =Employee("sdf",47,'m')
print(employee._name)

```

```python
## Encapsulation with getter and setter 
class Person:
    def __init__ (self,name,age):
        self.__name=name      # private access
        self.__age=age 

    #get method 
    def get_name(self):
        return self.__name
    
    # set method to change any values that u give 
    def set_name(self,name):
        self.__name=name

    # get for age
    def get_age(self):
        return self.__age
    
    # set for age
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age cant be negative")

    
per=Person("hel",87)
print(per.get_name())

print(per.get_age())

per.set_age(78)
print(per.get_age())

per.set_age(-8)
print(per.get_age())
```
