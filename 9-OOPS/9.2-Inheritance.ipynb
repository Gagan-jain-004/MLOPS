# Inheritance 

```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    
    def drive(self):
        print(f"the person will drive the {self.enginetype} ")

car1=Car(4,4,"petrol")
car1.drive()        
```

```python
# Single Inheritance
# like if a new car has Car class properties and also some additional propts so we inherit Car class on that by adding the additional ones also

class Tesla(Car):                    # Tesla inherting Car class
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)              # super to call the parent class thing  like here calling parent clas constructor
        self.is_selfdriving=is_selfdriving

    def selfdriving(self):
        print(f"Tesla supports self drving: {self.is_selfdriving}")

tesla1= Tesla(4,3,"electric",True)
tesla1.selfdriving()
tesla1.drive()
```

```python
## MUltiple Inheritance
# when a class inherits from more than one base class

# base class 1
class Animal:
    def __init__(self,name):
        self.name=name

    def speak(slef):
        print("Subclass must implement this method")

# base class 2 
class Pet:
    def __init__(self,owner):
        self.owner=owner 


# Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} say woof"

## create an object

doggy = Dog("Bud","Omni")
print(doggy.speak())
print(f"Owner: {doggy.owner}")
```
