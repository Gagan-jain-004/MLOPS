Polymorphism allows objects of different classes to  be treated as objects of a common superclass. It is achieved through method overriding and inerfaces

# Method Overriding

It allows a child class to provide a specific implementation of a method that is already defined in its parent class

```python
# Base class
class Animal:
    def speak(self):
        print(f"Base class")
    
# Derived class
class Dog(Animal):
    def speak(self):
        print(f"Dog barks")

d=Animal()
e=Dog()
d.speak()
e.speak()
```

```python
### Polymorphism with fxns and methods

class Shape:
    def area(self):
        return "The area of figure"
    
# Derived class1 
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height= height

    def area(self):
        return self.width*self.height
    
# Derived class 2

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14*self.radius * self.radius
    
## Fxn to demonstrate polymorphism

def printarea(shape):
    print(f"the area is {shape.area()}")

circle=Circle(1)
printarea(circle)
```

## Polymorphism with Abstract Base Classes(i.e. interfaces)
Abstract base classes(ABCs) are used to define common methods for a group of related objects .
abstract class defines fxn only not definition init

```python
from abc import ABC,abstractmethod

## Define an abstract class

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass                 # as it is a empty fxn toh pass likh do

# Derived class 1 
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
class Bike(Vehicle):
    def start_engine(self):
        return "Bike engine started"
    

# fxn to demonstrate polymorphism

def vehicle(v):
    return v.start_engine()

bike=Bike()    
print(vehicle(bike))    

car=Car()    
print(vehicle(car))    
```
