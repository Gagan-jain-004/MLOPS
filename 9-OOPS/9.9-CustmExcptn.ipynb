## Custom exception (raise and throw an exception)

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year= int(input("Enter the dob"))
age=2026-year

try:
    if age<=30 and age>=20:
        print("Age is valid so u can apply")
    else:
        raise dobException        # as u raise so it meant to be catch so use try except block used 
except dobException:
    print("sry,ur age should be > 20 and <30")
```
