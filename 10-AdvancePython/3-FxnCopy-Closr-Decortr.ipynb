Decorators is a flexible feature that allows u to modify the behaviour  of a fxn or class method. They r commonly used to add fxnality to fxns or methods without modifying their actual code.

```python
## Function copy
def welcome():
    return "Welcome to the program!"

welcome()
```

```python
# copying
wel=welcome     # copying welcome to wel
print(wel())
del welcome   # deleting original function
print(wel())
```

```python
## Closures -  fxn ke andr fxn

def mainwelcome():
    msg="welton"
    def subwelcomemethod():
        print("Inside adv python")
        print(msg)
        print("Sub wecm")
    return subwelcomemethod()

mainwelcome()
```

```python
## Closures -  taking fxn as argument also

def mainwelcome(func):
    def subwelcomemethod():
        print("Inside adv python")
        func("Hell yeah")
        print("Sub wecm")
    return subwelcomemethod()

mainwelcome(print)        # as print is inbuild fxn we can pass it directly
```

```python
# to print length of list

def mainwelcome(func,lst):
    def subwelcomemethod():
        print("Inside adv python")
        print(f"Length of list is: {func(lst)}")
        print("Sub wecm")
    return subwelcomemethod()

mainwelcome(len,[1,2,3,4,5])        # passing len fxn and list
```

```python
## Decorators

def mainwelcome(func):
    def subwelcomemethod():
        print("Inside adv python")
        func()
        print("Sub wecm")
    return subwelcomemethod()
```

```python
def courseintro():
    print("Welcome to advanced python course")

courseintro()
```

```python
mainwelcome(courseintro)    # our decorated fxn is use to get output of below withut calling mainwelcome fxn
```

```python
@mainwelcome                   #decorator keyword with @  so whenever we write this then below fxn act as argument to mainwelcome
def courseintro():
    print("Welcome to advanced python course")
```

```python
## Decorators

def mydecorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@mydecorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)     
def greet():
    print(f"Hello!")    
```

```python
greet()        # when it call repeat takes 3 and fxn greet as argument and calls wrapper which calls greet 3 times
```
