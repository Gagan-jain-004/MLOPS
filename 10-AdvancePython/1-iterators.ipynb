# Iterators
These are advanced python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
mylist=[1,2,3,4,5,6]
for i in mylist:
    print(i)
type(mylist)
print(mylist)
```

```python
## Iterator using iter keyword
iterator=iter(mylist)
print(type(iterator))
```

```python
# Iterate through all element
next(iterator)           # the more time u run it will go ahead ahead to print next index
```
