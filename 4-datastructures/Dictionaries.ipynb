Dictonaries are unordered collections of items . They store data in key-value pairs . keys must be unique and immutable ( e.g. strings,numbers or tuples) while values can be of any type 


```python
## Creating Dictionaries (created by {})
empty_dict={}
print(type(empty_dict)) 
```

```python
# Another way of creating dictionary using dict()
empty_dict = dict()
empty_dict
```

```python
student={"name":"rehman dakait","age":42,"grade":50}
print(student)
print(type(student))
```

```python
#In output only Single key only get printed
student={"name":"rehman dakait","age":42,"name":50}
print(student)
```

```python
# Accessing Dictionary elements 
student={"name":"rehman dakait","age":42,"grade":50}
print(student["grade"])

print(student.get("age"))
print(student.get("last_name"))
# giving default value  for absent key
print(student.get("last_name","Not Available"))


```

```python
# Modifying Dictionary Elements , Dictionary are mutable
student["age"]=12        ##update value for the key 
print(student)
student["address"]="India"          #added a new key and value 
print(student)
```

```python
del student['grade']    ## delete key and value pair
print(student)
```

```python
# Dictionary Methods
keys=student.keys()   # get all keys
print(keys)
values = student.values()  # get all values

items=student.items()    #get all key value pairs
print(items)
```

```python
# Copying  but on changes on top passes to copy also so for that shallow copy used 
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"]="rehman2"
print(student)
print(student_copy)
```

```python
# Shallow Copy 
student_copy1=student.copy()
print(student_copy1)
print(student)

student["name"] = "rehman3"
print(student_copy1)
print(student)
```

```python
# Iterating Over Dictionaries 
# Iterating over keys 

for keys in student.keys():
    print(keys)
```

```python
# Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")                    # f for formatting string
```

```python
## Nested Dictionaries
students={
    "student1":{"name":"rehman","age":42},
    "student2":{"name":"dakait","age":32}
}
print(students)
```

```python
#Access nested dictionary elements
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over Nested Dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
#Dictionary Comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
# Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
#PRACTICE QUESTIONS
# Use a dictionary to count the frequency of elements in a list

numbers=[1,2,2,3,4,4,4,5]
frequency={}

for num in numbers:
    if num in frequency:
        frequency[num]+=1
    else:
        frequency[num]=1
print(frequency)
```

```python
# Merge two dictionaries into one
dict1={"a":1,"b":2}
dict2={"c":3,"d":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
