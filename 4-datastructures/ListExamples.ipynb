EX 1: Create a todo list to keep track of tasks

```python
todolist=["buy groceries", "clean the house", "pay bills"]

# ADDING TO TASK
todolist.append("walk the dog")
todolist.append("read a book")

# REMOVING A TASK
todolist.remove("clean the house")

# Checking if a task is in the list
if "pay bills" in todolist:
    print("You have a bill to pay!")

print("to do list remaining")
for task in todolist:
    print(f"-{task}") 

```

Example 2: Organizing Student Grades
  -  Create a list to store and calculate average grades for students

```python
# Organizing student grades
grades =[85, 92, 78, 90, 88]

# Adding a new grade
grades.append(95)

# Calculating average grade
average_grade = sum(grades) / len(grades)
print(f"Average grade: {average_grade:.2f}")

# Finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
```
