# File operation- Read and write files

```python
## Read a whole file

with open('example.txt','r') as file:           # opening a file in read mode
    content=file.read()         # read to all content
    print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:           # opening a file in read mode
    for line in file:
        print(line)
```

```python
## Wtiting(overwriting completely) a file

with open('example.txt','w') as file:       # w - write mode so it replace with below content
    file.write('Hll well\n')
    file.write('this is a new line')
```

```python
## Write a file without overwriting 

with open('example.txt','a') as file:          # a - append mode 
    file.write("Appending content in a file without overwriting")
```

```python
## writing a list of lines to a file
lines=['First line \n','Second line \n','thirdline \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# Binarry Files  (with .bin extension and dealing with bytes happen in it)

# writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin','rb')as file:
    content =file.read()
    print(content)
```

```python
### Read the content from source text file and write to destination text file

#copying a text file
with open('example.txt','r') as src_file:
    content=src_file.read()

with open('destination.txt','w') as dest_file:
    dest_file.write(content)
```

```python
# Read a text file nd count the number of lines, words, characters

def count_text_file(filepath):
    with open(filepath,'r') as file:
        lines= file.readlines()
        linecount=len(lines)
        wordcount=sum(len(line.split()) for line in lines)
        charcount=sum(len(line) for line in lines)
    return linecount,wordcount,charcount

filepath='example.txt'
lines,words,chars=count_text_file(filepath)
print(f"Lines: {lines}, Words: {words}, Characters: {chars}")
```

The W+ mode in python is used to open a file for both reading and writing . if the file doesn't exist , it will be created. if the file exists, its content is truncated(i.e the file is overwritten)

```python
## writing and then reading a file

with open('example.txt','w+')as file:
    file.write("helll oooooo\n")
    file.write("this is second line\n")

    # move the cursor to the beginning of the file as cursor is at the end after writing
    file.seek(0)

    # read the content of the file
    content=file.read()
    print(content)
```
