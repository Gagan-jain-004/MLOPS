```python
# Series
## A pandas Series is a one-dimensional array-like object that can hold any data type.
#  It is similar to a column in a spreadsheet or a database table.

import pandas as pd
data=[1,2,3,4,5]
series = pd.Series(data)
print("Series \n",series)  # in output the left sides are indexes not wry abt them

print(type(series))
```

```python
# dictionaries to series
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)     # u will see that key becomes the index for it 
```

```python
# another way to create series
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
pd.Series(data, index=index)
```

```python
## DataFrame is just like a table with rows and columns.

# create a DataFrame from a dictionary of list 

data={
    'Name':['Kri','joh','jac'],
    'Age':[25,30,45],
    'City':['bangalore','Delhi','jaipur']
}
df=pd.DataFrame(data)
print(df)
```

```python
# create a dataframe from list[] of dictionaries{}

data=[
     {'Name':'Tris','Age':78,'City':"Bang"},
     {'Name':'Nicohl','Age':78,'City':"Bang"},
     {'Name':'rock','Age':78,'City':"Bang"},
     {'Name':'Jac','Age':78,'City':"Bang"},
]

df=pd.DataFrame(data)
print(df)
```
