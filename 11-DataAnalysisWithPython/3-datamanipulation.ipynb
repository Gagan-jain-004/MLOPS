## Data manipulation and analysis with pandas
these are the key tasks in any datascience or data analyst project. Pandas provides wide range of fxns for it , making it easier to clean ,tranform and extrct insights from data. 

```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data.csv')
# fetch the first 5 rows
df.head(5)
```

```python
df.tail(5)
```

```python
df.describe()      # to get statistical summary of numerical columns
```

```python
df.dtypes
```

```python
## Handling Missing Values
df.isnull().any()         # agr true hoga to column me missing values hain

```

```python
df.isnull().sum()      # har column me kitni missing values hain
```

```python
## filling missing values with mean (for numerical columns)
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Columns
df = df.rename(columns={'Order Date':'Sale Date'})
df.head()
```

```python
groupedmean=df.groupby('Item Type')['Unit Price'].mean()
print(groupedmean)
```

```python
groupedsum=df.groupby(['Item Type','Region'])['Unit Price'].sum()
print(groupedsum)
```

```python
## aggregate multiple fxns
groupedagg=df.groupby('Region')['Unit Price'].agg(['mean','sum','count'])
groupedagg
```

```python
## Merging and joining Dataframes
# Create sample dataframes 
df1=pd.DataFrame({'key':['A','B','c'],'Value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge dataframe on the key column
pd.merge(df1,df2,on='key',how="inner")
```

```python
pd.merge(df1,df2,on='key',how="outer")
```

```python
pd.merge(df1,df2,on='key',how="left")
```

```python
pd.merge(df1,df2,on='key',how="right")
```
